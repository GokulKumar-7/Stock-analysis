```python
import numpy as np
import pandas as pd
import os

df = pd.read_csv('/content/creditcard_2023.csv')
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df
```

```python
df.describe()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
q1,q3
```

```python
iqr = q3 - q1
iqr
```

```python
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr
lower_limit, upper_limit
```

```python
outliers = df[(df < lower_limit) | (df > upper_limit)].any(axis=1)

df[outliers]
```

```python
df_no_outliers = df[(df > lower_limit) & (df < upper_limit)].any(axis=1)
filtered_df = df[df_no_outliers]

# Convert the datatype of 'V1' column to float64
filtered_df['V1'] = filtered_df['V1'].astype('float64')

# Access the column (now with float64 datatype)
filtered_df
```

```python
from sklearn.model_selection import train_test_split
filtered_df
```

```python
X = filtered_df.drop(columns = ['Class'])
y = filtered_df['Class']
```

```python
X.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 35, test_size =0.35)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```
